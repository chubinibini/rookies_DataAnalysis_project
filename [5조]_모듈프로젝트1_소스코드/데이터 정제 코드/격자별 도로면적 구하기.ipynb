# 라이브러리 및 경로 설정

```python
import geopandas as gpd
import pandas as pd
from shapely import wkt
```

```python
src_path = '/content/drive/MyDrive/sk쉴더스/프로젝트1/데이터/'
```

# 데이터 로드

```python
road = gpd.read_file(src_path + '실폭도로/Z_KAIS_TL_SPRD_RW_11000.shp')
```

```python
road.crs
```

```python
road = road.to_crs('epsg:5179')
```

```python
grid = pd.read_csv(src_path + 'unique_grid.csv')
```

```python
grid['geometry'] = grid['geometry'].apply(wkt.loads)
geo_grid = gpd.GeoDataFrame(grid, crs='epsg:5179')
```

```python
geo_grid
```

# 서울 모든 도로 합집합

```python
seoul_road = road.buffer(0.1).unary_union
```

```python
seoul_road.area
```

# 도로와 격자 교집합 면적 구하기

## 하나만 예시

```python
geo_grid[geo_grid.index == 0].explore()
```

```python
geo_grid[geo_grid.index == 0].intersection(seoul_road).explore()
```

```python
geo_grid['도로면적'] = 0
for i in geo_grid.index:
  geo_grid.loc[geo_grid.index == i, '도로면적'] = geo_grid[geo_grid.index == i].intersection(seoul_road).area
```

```python
geo_grid.to_csv(src_path + 'road_area.csv', index=False)
```
