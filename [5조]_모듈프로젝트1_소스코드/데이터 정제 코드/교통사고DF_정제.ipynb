## 결측치 확인 및 채우기

- 빈 값이 24315개, 전체의 2퍼센트
- 가해 및 피해 운전자 연령에 미분류 값이 존재
- 결측 존재 컬럼
  - acdnt_age_2_dc(피해 운전자 연령) :  912개의 NaN존재 -> 공작물충돌, 전도전복 등 사유로 차량 단독 사고 경우 존재 -> '없음'으로 대체
    - dmge_vhcle_asort_dc(피해 운전자 차종)
      - acdnt_age_2_dc와 마찬가지 사유로 NaN 존재 -> '없음'으로 대체
    - injury_dgree_2_dc(피해 운전자 상해정도)
      - acdnt_age_2_dc와 마찬가지 사유로 NaN 존재 -> '없음'으로 대체
    - sexdstn_div_2_dc(피해 운전자 성별)
      - acdnt_age_2_dc와 마찬가지 사유로 NaN 존재 -> '없음'으로 대체
  - acdnt_pos(정확한 사고 위치) : 11987개(전체 35.3%) 빈 값 존재 -> 빈값을 '기록없음'으로 대체
  - route_nm(사건발생 위치:도로명) : 4336(전체 12.8%)의 결측치 -> '기록없음'로 대체
    - road_div : 4336(전체 12.8%)의 결측치 -> route_nm에 결측이 있으면 road_div에도 결측이 있다 -> 결측을 같은 값으로 채워줌
- bdy_injury_part_1_dc, bdy_injury_part_2_dc는 3개의 결측이 존재 -> 상해 부위가 존재하지 않는 것이 아닌 기록이 되지 않은 것을로 보임
  - 상해 정도에는 중상으로 들어가 있으나, bdy_injury_part가 결측치인 것을 통해 추측
  - 양 쪽 모두 '기록없음'으로 결측치를 기록


```python
# acdnt_age_2_dc(피해 운전자 연령)에 NaN이 값이 들어있는 컬럼 확인
# -> 공작물충돌, 전도전복 등 차량 단독 사고인 경우 존재 -> 빈값을 '없음'으로 대체
# dmge_vhcle_asort_dc, injury_dgree_2_dc, sexdstn_div_2_dc도 마찬가지
acci_df['acdnt_age_2_code'] = acci_df['acdnt_age_2_code'].fillna(1000)
acci_df['dmge_vhcle_asort_dc'] = acci_df['dmge_vhcle_asort_dc'].fillna('없음')
acci_df['injury_dgree_2_dc'] = acci_df['injury_dgree_2_dc'].fillna('없음')
acci_df['sexdstn_div_2_dc'] = acci_df['sexdstn_div_2_dc'].fillna('없음')
```

```python
# acdnt_pos(정확한 사고 위치)에 다수의 빈 값 존재 -> 빈값을 미식별로 대체
acci_df['acdnt_pos'] = acci_df['acdnt_pos'].fillna('기록없음')
```

```python
# NaN을 가진 컬럼끼리 특별한 연관성은 안보이므로 일단 기록없음으로 대체
acci_df['route_nm'] = acci_df['route_nm'].fillna('기록없음')
acci_df['road_div'] = acci_df['road_div'].fillna('기록없음')
```

```python
acci_df['bdy_injury_part_1_dc'] = acci_df['bdy_injury_part_1_dc'].fillna('기록없음')
acci_df['bdy_injury_part_2_dc'] = acci_df['bdy_injury_part_2_dc'].fillna('기록없음')
```
