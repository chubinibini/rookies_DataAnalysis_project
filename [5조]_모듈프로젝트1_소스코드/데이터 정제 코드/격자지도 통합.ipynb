# 자치구 이름 정렬

```python
import glob

shps = glob.glob('/content/drive/MyDrive/프로젝트1/데이터/격자지도/*/*.shp')
shps.sort()
shps
```

# 격자지도 파일 통합하여 하나의 데이터 프레임 생성

```python
dfs = pd.DataFrame()
for shp in shps:
  print( shp )
  gu = shp.split('/')[7]
  df = gpd.read_file(shp, encoding='utf-8')
  df['gu'] =  gu
  dfs = pd.concat( [dfs, df] )
  display( dfs ) 
  #break
```
