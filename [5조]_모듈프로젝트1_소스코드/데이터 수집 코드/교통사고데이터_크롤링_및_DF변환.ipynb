# 데이터 크롤링

- 수업에서 배운 내용이 아닌 사이트 내 Devtools에서 직접 긁어오는 방식 채용
- GET방식이 아닌 POST방식으로 전송하는 사이트임을 고려하여 크롤링 진행

```python
url = 'http://taas.koroad.or.kr/gis/srh/ash/selectAccidentInfo.do'

payload = {
'acdntGaeCode':"01,02,03,04",
'endAcdntYear':"2021",
'engnCode':"00",
'legaldongCode':"11%",
'pageIndex':1,
'searchType':"00",
'startAcdntYear':"2021",
'zoneYn':False
}
response1 = requests.post(url,json=payload)
dict = json.loads(response1.text)
# 서울 전체 기준 교통사고 데이터 추출 -> {  [ {  }, {  }, {  }, ... ]  }형태
```
